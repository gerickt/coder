### I: Construcción de los dataframes

```python
# Librerias que vamos a usar
import requests
import pandas as pd
```

```python
# Configuración de la API
api_key = "XXXXXXXXXXX"  # <- removida intencionalmente
buscar_keyword = "%incendios%"
barrido = 0

# Endpoint y Parámetros
url = "https://api.gerick.dev/api/rest/centinelanews"
params = {"keyword": buscar_keyword, "offset": barrido}

# Headers
headers = {"x-secret": api_key}
```

```python
# Hacer la Solicitud
response = requests.get(url, params=params, headers=headers)
data = response.json()

# Convertir a DataFrame la lista 'vista_rss_6meses'
df_rss = pd.DataFrame(data["vista_rss_6meses"])
# Convertir a DataFrame la lista 'vista_spider_6meses'
df_spider = pd.DataFrame(data["vista_spider_6meses"])

# Concatenar DataFrames
df_combinado = pd.concat([df_rss, df_spider], ignore_index=True)
```

```python
# limpiando los resultados
from bs4 import BeautifulSoup


def limpiar_html(texto):
    # Analiza el texto con BeautifulSoup y obtén solo el texto
    soup = BeautifulSoup(texto, "html.parser")
    return soup.get_text()


# Aplicar la función de limpieza a cada celda que contenga HTML
df_combinado["content"] = df_combinado["content"].apply(limpiar_html)
```
