```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

df_ejercicio = sns.load_dataset("exercise")
df_ejercicio = df_ejercicio.drop("Unnamed: 0", axis="columns")
print(df_ejercicio.head())
```

```python
df_30_min = df_ejercicio[df_ejercicio["time"] == "30 min"]
print(df_30_min.head())
```

```python
plt.figure()
sns.displot(data=df_30_min, x="pulse", kind="kde", hue="kind", fill=True)
```

```python
plt.figure()
sns.catplot(
    data=df_ejercicio,
    kind="violin",
    x="time",
    y="pulse",
    hue="diet",
    split=True,
    observed=True,
)
```
