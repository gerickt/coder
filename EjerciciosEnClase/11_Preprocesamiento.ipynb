# Outliers

```python
# Metodo 1: IQR
primer_cuartil = variable_target.quantile(0.25)
tercer_cuartil = variable_target.quantile(0.75)
rango_intercuartilico = tercer_cuartil - primer_cuartil

# Limites:
limite_inf = primer_cuartil - (1.5 * rango_intercuartilico)
limite_sup = tercer_cuartil + (1.5 * rango_intercuartilico)

print(limite_inf, limite_sup)
```

```python
# Identifico outliers:
df["Otlier_IQR"] = np.where(
    (df["SalePrice"] < limite_inf) | (df["SalePrice"] > limite_sup), 1, 0
)
df["Otlier_IQR"].value_counts()
```

```python
# Grafico
plt.figure()
sns.scatterplot(data=df, x="Id", y="SalePrice", hue="Otlier_IQR")
plt.axhline(
    y=limite_inf, color="black", linestyle="--", label="Limite Inferior", linewidth=2
)
plt.axhline(
    y=limite_sup, color="black", linestyle="--", label="Limite Superior", linewidth=2
)
plt.title("DetecciÃ³n Outliers - Metodo IQR")
```

```python
# Metodo 2: Z-Score
media = variable_target.mean()
desvio = variable_target.std()
valores_z = (variable_target - media) / desvio

# Limites:
limite_inf = -3
limite_sup = 3

print(valores_z)
```

```python
# Identifico outliers:
df["Otlier_ZScore"] = np.where(
    (valores_z < limite_inf) | (valores_z > limite_sup), 1, 0
)
df["Otlier_ZScore"].value_counts()
```

```python
limite_inf_tr = limite_inf * desvio + media
limite_sup_tr = limite_sup * desvio + media

# Grafico
plt.figure()
sns.scatterplot(data=df, x="Id", y="SalePrice", hue="Otlier_ZScore")
plt.axhline(
    y=limite_inf_tr, color="black", linestyle="--", label="Limite Inferior", linewidth=2
)
plt.axhline(
    y=limite_sup_tr, color="black", linestyle="--", label="Limite Superior", linewidth=2
)
plt.title("DetecciÃ³n Outliers - Metodo Z Score")
```

```python
# Metodo 3: Distancia Media
media = variable_target.mean()
MAD = sum(abs(variable_target - media)) / variable_target.count()

# Limites:
limite_inf = media - (3 * MAD)
limite_sup = media + (3 * MAD)

print(MAD, limite_inf, limite_sup)
```

```python
# Identifico outliers:
df["Otlier_DistanciaMedia"] = np.where(
    (df["SalePrice"] < limite_inf) | (df["SalePrice"] > limite_sup), 1, 0
)
df["Otlier_DistanciaMedia"].value_counts()
```

```python
# Grafico
plt.figure()
sns.scatterplot(data=df, x="Id", y="SalePrice", hue="Otlier_DistanciaMedia")
plt.axhline(
    y=limite_inf, color="black", linestyle="--", label="Limite Inferior", linewidth=2
)
plt.axhline(
    y=limite_sup, color="black", linestyle="--", label="Limite Superior", linewidth=2
)
plt.title("DetecciÃ³n Outliers - Metodo Distancia de la Media")
```

# Outliers: AnÃ¡lisis Multivariado

### GarageArea: 
Size of garage in square feet. 

```python
# CorrelaciÃ³n
df[["GarageArea", "SalePrice"]].corr(method="pearson")
```

```python
# Metodo 3: Distancia Media Multivariado
# Defino funciÃ³n:
def outlier_euclideano_d(x, cutoff):
    # x: dataframe con valors numericos normalizados
    result_ = pd.Series([0] * len(x.iloc[:, 1]))
    data_mean = x.mean()  # media de los datos_
    dist = np.sqrt(np.sum(((x - data_mean) ** 2), axis=1))  # Distancia euclideana
    dist_mean = dist.mean()  # media de las distancia
    dist_zscore = np.abs((dist - dist_mean) / dist.std())  # z-score para las distancias
    result_[((dist_zscore > 3))] = 1
    return result_
```

```python
# Aplicar la funcion
df["Otlier_DistanciaMedia_Mult"] = outlier_euclideano_d(
    df[["GarageArea", "SalePrice"]], 3
)
df.head()
```

```python
# Grafico
plt.figure()
sns.scatterplot(
    data=df, x="GarageArea", y="SalePrice", hue="Otlier_DistanciaMedia_Mult"
)
plt.title("DetecciÃ³n Outliers - Metodo Distancia de la Media - Multivariado")
```
